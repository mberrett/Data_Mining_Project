```python
import numpy as np
import matplotlib.pyplot as plt

N = 3
# These score will be in the following oder -
# DropNA, Mode, KNN, RandomForest 

# Before Normalization 
Accuracy_score = (81.69, 85.41, 85.51)
Precision_score = (61.67, 59.75, 64.12)
Recall_score = (61.15, 73.51, 71.58)
F1_score = (61.41, 65.92, 67.65)

'''
# After Normalization 
Accuracy_score = (86.86, 81.95, 81.97)
Precision_score = (32.48, 32.71, 32.45)
Recall_score = (77.82, 78.23, 78.69)
F1_score = (45.83, 46.13, 45.95)

# Parametre Optimized 
Accuracy_score = (82.51, 85.98, 85.73)
Precision_score = (62.38, 58.16, 63.65)
Recall_score = (63.15, 76.85, 72.55)
F1_score = (62.76, 66.21, 67.81)

# Feature Selection 
Accuracy_score = (82.51, 84.91, 85.73)
Precision_score = (62.38, 68.54, 63.65)
Recall_score = (63.15, 67.89, 72.55)
F1_score = (62.76, 68.21, 67.81)

# Improvements
Accuracy_score = (82.51, 84.91, 85.73)
Precision_score = (62.38, 68.54, 63.65)
Recall_score = (63.15, 67.89, 72.55)
F1_score = (62.76, 68.21, 67.81)
'''

ind = np.arange(N, dtype='float')  # the x locations for the groups
width = 0.15      # the width of the bars

fig, ax = plt.subplots(figsize=(12,12))
rects1 = ax.bar(ind, Accuracy_score, width, color='g')
rects2 = ax.bar(ind + width, Precision_score, width, color='y')
rects3 = ax.bar(ind + 2*width, Recall_score, width, color='blue')
rects4 = ax.bar(ind + 3*width, F1_score, width, color='orange')

# add some text for labels, title and axes ticks
ax.set_ylabel('Percentage Score')
ax.set_title('Model Performance by Imputation Techniques')
ax.set_xticks(ind + width*1.5)
ax.set_xticklabels(('Mode', 'KNN', 'RandomForest'))

ax.legend((rects1[0], rects2[0], rects3[0], rects4[0]), ('Accuracy', 'Precision', 'Recall', 'F1-Score'))


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

plt.show()
```
